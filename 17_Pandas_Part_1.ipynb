# Creating Dataframes

1. What is a dataframe?
2. How is it related to other data storage options?

A DataFrame is a fundamental data structure in the Python programming library pandas, which is designed for data manipulation and analysis. Essentially, a DataFrame represents a table of data with rows and columns, similar to a spreadsheet, a SQL table, or data in a CSV file. Each column in a DataFrame can hold data of a single data type, but different columns can hold different types of data (e.g., integers, floats, strings, Python objects, etc.).

# Fixing Null Values

```python
import pandas as pd
import numpy as np

# Creating a small dummy DataFrame with null values
np.random.seed(0)  # For reproducible output
data = {
    'A': [1, 2, np.nan, 4, 5, np.nan, 7, 8, 9, 10],
    'B': [np.nan, 1, 2, 3, 4, 5, np.nan, 7, 8, 9],
    'C': [2, 3, 4, np.nan, np.nan, np.nan, 7, 8, 9, 10],
    'D': ['x', np.nan, 'y', 'x', 'y', 'x', np.nan, 'y', 'x', 'y']
}

df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
df
```

```python
|# Dropping all values of Null in df
df.dropna()  # Drops rows with any null values
```

```python
df.dropna(how='any')  # Drops rows where all values are null
```

```python
# Drop columns with any null values
df.dropna(axis=1)
```

```python
# Fill with a specific value
df.fillna(0)
```

```python
# Fill with mean for numeric columns
df.fillna(df.mean())
```

```python
df.columns
```

```python
df
```

```python
# Correctly fill NaN values in columns 'A' and 'C' with 0
df[['A', 'C']] = df[['A', 'C']].fillna(0)

# Fill NaN values in column 'B' with its own mean
df['B'] = df['B'].fillna(df['B'].mean())

# Fill NaN values in column 'D' with its mode
# This assumes df['D'].mode()[0] does not result in an empty series; otherwise, you should handle that case.
df['D'] = df['D'].fillna(df['D'].mode()[0])

df
```

```python
# Fill categorical data with the mode
df.fillna(df['D'].mode()[0])
```

# Group By In DataFrames

```python
import pandas as pd

# Creating the first DataFrame: Orders
data1 = {
    'OrderID': [1, 2, 3],
    'CustomerName': ['Alice', 'Bob', 'Charlie']
}
df1 = pd.DataFrame(data1)

print(df1)

# Creating the second DataFrame: Items in Orders
data2 = {
    'OrderID': [1, 1, 2, 2, 3, 3, 3],
    'ItemName': ['ItemA', 'ItemB', 'ItemB', 'ItemC', 'ItemA', 'ItemB', 'ItemC']
}
df2 = pd.DataFrame(data2)

print(df2)

# Performing an inner join on the 'OrderID' column
result = pd.merge(df1, df2, on='OrderID')

print(result)
```
